## Import csv

```python
import os
import glob
import pandas as pd

dataset = pd.read_csv('./HTR ML - height.csv')
```

```python
dataset.describe()
```

## Create TGAN instance

```python
from tgan.model import TGANModel
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

continuous_columns = [0, 1, 2]

tgan = TGANModel(
        continuous_columns,
        output='output',
        max_epoch=5,
        steps_per_epoch=10000,
        save_checkpoints=True,
        restore_session=True,
        batch_size=50,
        z_dim=200,
        noise=0.2,
        l2norm=0.00001,
        learning_rate=0.001,
        num_gen_rnn=100,
        num_gen_feature=100,
        num_dis_layers=1,
        num_dis_hidden=100,
        optimizer='AdamOptimizer'
    )

tgan.fit(dataset)
```

```python
num_samples = 1000
samples = tgan.sample(num_samples)
samples.describe()
```

```python
samples.to_csv('Generated_data.csv')
```
